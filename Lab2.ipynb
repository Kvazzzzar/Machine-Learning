# Расчет теста Пирсона
Коэффициент корреляции Пирсона характеризует существование линейной зависимости между двумя величинами.

```python
df_not_null[df_not_null.select_dtypes(include=[np.number]).columns].corr()
```

Этот вывод — это **матрица корреляции**, которая показывает, как связаны между собой различные переменные. Корреляция может варьироваться от -1 до 1:
- **1** означает, что переменные полностью положительно коррелируют (если одна увеличивается, другая тоже увеличивается).
- **-1** означает, что переменные полностью отрицательно коррелируют (если одна увеличивается, другая уменьшается).
- **0** означает отсутствие линейной связи.

### Пример разбора:

1. **symboling**:
   - Корреляция с **wheel-base**: -0.54. Это означает, что чем больше колесная база, тем меньше риск по оценке символов.
   - Корреляция с **price**: -0.09. Связь между символингом и ценой слабая и отрицательная, что означает, что небольшие изменения в символинге слабо связаны с уменьшением цены.

2. **normalized-losses**:
   - Корреляция с **price**: 0.13. Это положительная, но слабая корреляция, что означает, что когда увеличиваются нормализованные потери, цена имеет тенденцию к небольшому увеличению.

3. **wheel-base**:
   - Корреляция с **length**: 0.88. Очень высокая положительная корреляция, что логично — чем больше длина автомобиля, тем больше у него колесная база.
   - Корреляция с **price**: 0.59. Увеличение колесной базы связано с увеличением цены.

4. **length**:
   - Корреляция с **width**: 0.86. Широкие автомобили, как правило, длиннее.
   - Корреляция с **price**: 0.69. Большие по длине автомобили имеют тенденцию быть более дорогими.

5. **width**:
   - Корреляция с **price**: 0.75. Широкие автомобили обычно дороже.

6. **curb-weight** (снаряжённый вес):
   - Корреляция с **price**: 0.84. Чем тяжелее автомобиль, тем выше его цена.

7. **engine-size**:
   - Корреляция с **price**: 0.87. Размер двигателя является одним из ключевых факторов, влияющих на цену.

8. **compression-ratio**:
   - Корреляция с **price**: 0.07. Слабая положительная корреляция, что говорит о слабой связи между степенью сжатия и ценой.

9. **city-mpg** и **highway-mpg**:
   - Корреляция между собой: 0.97. Это очень логично, так как автомобили, которые эффективны в городе, также, как правило, эффективны на шоссе.
   - Оба параметра имеют отрицательную корреляцию с **price**: чем выше расход топлива (в милях на галлон), тем ниже цена.

### Основные выводы:
- Цена автомобиля больше всего коррелирует с такими переменными, как **engine-size** (0.87), **curb-weight** (0.84), и **width** (0.75).
- Такие переменные, как **symboling** и **height**, имеют слабую или почти нейтральную корреляцию с ценой.

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
```

```python

# Выбираем категориальные признаки
categorical_features = X.select_dtypes(include=['object', 'int64']).columns.tolist()

print("Категориальные признаки:", categorical_features)
```

```python
# Создаем копию данных
X_processed = X.copy()

# Применяем One-Hot Encoding
X_processed = pd.get_dummies(X_processed, columns=categorical_features, drop_first=True) # drop_first=True позволяет избежать мультиколлинеарности, удаляя первый уровень категориальной переменной.
```

```python
X_processed
```

```python
# Выбираем числовые признаки
numeric_features = X_processed.select_dtypes(include=['float64']).columns.tolist()

print("Числовые признаки:", numeric_features)
```

```python

# Инициализируем scaler
scaler = MinMaxScaler()

# Применяем нормализацию
X_processed[numeric_features] = scaler.fit_transform(X_processed[numeric_features])
```

```python
X_processed
```

```python
def train_and_evaluate(X, Y):
    # Разделение данных на обучающую и тестовую выборки
    X_train, X_test, Y_train, Y_test = train_test_split(
        X, Y, test_size=0.2, random_state=42, stratify=Y)

    # Инициализация модели
    model = LogisticRegression(max_iter=1000)

    # Обучение модели
    model.fit(X_train, Y_train)

    # Предсказания на обучающей выборке
    Y_train_pred = model.predict(X_train)
    train_accuracy = accuracy_score(Y_train, Y_train_pred)

    # Предсказания на тестовой выборке
    Y_test_pred = model.predict(X_test)
    test_accuracy = accuracy_score(Y_test, Y_test_pred)

    # Вывод результатов
    print(f"Точность на обучающей выборке: {train_accuracy:.4f}")
    print(f"Точность на тестовой выборке: {test_accuracy:.4f}")

    # Классификационный отчет
    print("\nКлассификационный отчет на тестовой выборке:")
    print(classification_report(Y_test, Y_test_pred))

    return model
```

```python
# Обучение модели
model = train_and_evaluate(X_processed, Y)
```

```python
# Получение коэффициентов модели
coefficients = pd.DataFrame({
    'Feature': X_processed.columns,
    'Coefficient': model.coef_[0]
})

# Сортировка по абсолютному значению коэффициента
coefficients['Abs_Coefficient'] = coefficients['Coefficient'].abs()
coefficients = coefficients.sort_values(by='Abs_Coefficient', ascending=False)

print(coefficients[['Feature', 'Coefficient']])
```
