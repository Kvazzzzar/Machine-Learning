# Визуализация распределения переменных

---



```python
sns.histplot(df['price'], kde = True)
plt.title( 'Распределение цен на машины:')
plt.xlabel('Цена $')
plt.ylabel('Кол-во')
plt.show()
```

```python
plt.figure(figsize=(25, 10))
sns.countplot(x='make', data=df)
plt.title('Марки авто')
plt.xlabel('Производитель')
plt.ylabel('Кол-во')
plt.show()
```

```python
sns.boxplot(x=df['engine-size'])
plt.title('Boxplot объем двигателя')
plt.xlabel('объем двигателя (литры)')
plt.show()
```

Проанализировав графики представленные выше (цена, марки авто, объем двигателя) можно прийти к следующим выводам:
1. Больше всего буджетных автомобилей с ценой 5-10к $
2. Наиболее популрна марка Toyota
3. Авто с объемом двигателя от 100 до 150 литров - самые распространнёные.

Итак, из п. 1 можно сделать закономерный вывод что бюджетные авто больше всего ценятся, так как чем больше охват рынка (покупательская способность + экспорт за границу - пошлины, транзит, диллеры) - тем выше прибыль у производителя.
Из п. 2 можно сделать вывод что Toyota - наиболее распространённая марка авто, это и понятно, это же Toyota - она не ломается,  не стоит забывать, что Toyota создательница таких творений как Mark II и SUPRA (s-u-u-upra).
П. 3 нам демонстрирует доминацию средне-литровых автомобилей - можно предположить, что это связано с целевый использованием авто - а именно передвижение по дорогам общего пользования (в пределах населённого пунка или между ними), на которых, по большому счёту, большая динамика и не нужна (BMW-воды искл.)
